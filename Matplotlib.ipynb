Matplotlib

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data


```python
# sin wave || line plot

plt.plot(x,y,color='teal')
plt.show()
```

```python
# cosin wave || Line Plot
plt.plot(x,z, color = 'green')
plt.show()
```

```python
# adding title x - axis and y - axis labels
plt.plot(x,y, color = 'purple')
plt.xlabel('Angle')
plt.ylabel('Sin Value')
plt.title('Sin wave')
plt.show()
```

```python
# PARABOLA
x = np.linspace(-10,10,20)  # gives equally shaped 20 values betn -10 & 10
y = x**2
plt.plot(x,y, color = 'black')
plt.show()
```

```python
# with diff symbols other than line
plt.plot(x,y,'r+') # r is red color, + is symbol
plt.show()
```

```python
plt.plot(x,y,'g.') # values are plotted in green with the symbols assigned
plt.show()
```

```python
plt.plot(x,y, 'bx')
plt.show()
```

```python
# plot multiple graphs in a single

x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x),'g-')
plt.plot(x, np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
# ax = axis
ax = fig.add_axes([0,0,1,1]) # 0s are origin, 1s are  height and width
languages = ['Nepali','English','French','Newari','German']
people = [243,200,18,122,53]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

PIE CHART


```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['Nepali','English','French','Newari','German']
people = [300,105,58,132,89]
ax.pie(people, labels = languages, autopct = '%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='Blue')
ax.scatter(x,z,color='Red')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')   #axes is rectangle by default so projection
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z, cmap='Blues')
plt.show()                 #spiral scatter plot
```
